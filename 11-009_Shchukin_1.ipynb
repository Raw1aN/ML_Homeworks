Дано:

изображение (формат jpg/png)
диапазон строк для маски (y_min, y_max)
диапазон столбцов для маски (x_min, x_max)
число разбиений на области (h, например, для h=10 маска разбивается на 10*10=100 равных областей)
Требуется:

написать алгоритм усреднения цвета в каждой области (усреднение по каждому цветовому каналу: R,G,B)
прислать файл(ы): исходное изображение, изображение с маской, код (для работы в формате ipynb - всё в одном файле, в противном случае - все файлы поместить в один архив)

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread

orig_img = imread('https://pm1.aminoapps.com/7614/fb62f569ae0215558294319d10fcb4a06cee7eefr1-1356-1078v2_hq.jpg')

plt.imshow(orig_img)
plt.show()
```

```python
x_min, x_max, y_min, y_max = 280, 720, 250, 400
h = 10

def makeCustomImage(x_min_image, x_max_image, y_min_image, y_max_image, h_count, image) :
  x_part = (x_max_image - x_min_image) // h_count
  y_part = (y_max_image - y_min_image) // h_count
  customImage = image.copy()
  for x in range(x_min_image, x_max_image, x_part):
        for y in range(y_min_image, y_max_image, y_part):
          customImage[y:y+y_part, x:x+x_part] = np.mean(image[y:y+y_part, x:x+x_part], axis=(0, 1))
  return customImage

customImage = makeCustomImage(x_min, x_max, y_min, y_max, h, orig_img)

plt.imshow(customImage)
plt.show()
```
