**1. Построить функцию линейной регрессии. Найти значение при $\alpha=20000$ с точностью двух знаков после запятой.**
$$
\begin{array}{c|c|c|c|c|c|c|c}
t & 0 & 100 & 150 & 200 & 250 & 300 & 360
\\
\hline
\alpha & 18179 & 18216 & 18261 & 18323 & 18403 & 18500 & 18641
\end{array}
$$

```python
# входные данные
x = np.array([[18179], [18216], [18261], [18323], [18403], [18500], [18641]])
y = np.array([0, 100, 150, 200, 250, 300, 360])
```

```python
# разбиение данных на тренировочные и тестовые данные
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state=0, shuffle=True)
```

```python
# обучение модели
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
## Вычислаяем значение
alpha = np.array([[20000]])
predict = round(float(regressor.predict(alpha)),2)
print("Значение при альфа 2000")
print(f"{predict}")
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Линейная регрессия')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.pipeline import Pipeline

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
columns = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors",
           "body-style", "drive-wheels", "engine-location", "wheel-base", "length", "width",
           "height", "curb-weight", "engine-type", "num-of-cylinders", "engine-size",
           "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm",
           "city-mpg", "highway-mpg", "price"]
df = pd.read_csv(url, names=columns, na_values="?")
```

```python
df.drop("make", axis=1, inplace=True)
df.dropna(inplace=True)

X = df[["horsepower"]]
y = df["price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state=0, shuffle=True)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Линейная регрессия')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
poly_features = PolynomialFeatures(degree=6)
X_poly = poly_features.fit_transform(X_train)
```

```python


poly_reg = Pipeline(steps=[('poly_features', PolynomialFeatures(degree=2)),
                             ('regressor', LinearRegression())])

poly_reg.fit(X_poly, y_train)
```

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, poly_reg.predict(X_poly), color = 'blue')
plt.title('Полиномная регрессия')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
